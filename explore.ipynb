```python
import json
import pandas as pd

# Putanja do generisanog fajla
file_path = "data/output/telemetry_logs.jsonl"

# Čitamo liniju po liniju, svaki batch je JSON
events = []
with open(file_path, "r") as f:
    for line in f:
        batch = json.loads(line)
        for log_event in batch["logEvents"]:
            # 'message' je string JSON, pa ga parsiramo
            event = json.loads(log_event["message"])
            events.append(event)

# Prebacimo u DataFrame za lakšu analizu
df = pd.DataFrame(events)
print(df.head())
```

```python
print("Ukupno eventova:", len(df))
```

```python
print(df['body'].value_counts())
```

```python
# Pogledaj sve kolone iz events
print(df.columns)

# Ako te zanimaju samo atribute iz 'attributes'
df_attributes = pd.json_normalize(df['attributes'])
print(df_attributes.head(50))
```

```python
# Broj unikatnih korisnika
print("Broj korisnika:", df_attributes['user.id'].nunique())

# Tipične vrednosti za model
print(df_attributes['model'].value_counts().head(10))
```

```python
print("Tipične vrednosti za tool:")
print(df_attributes['tool_name'].value_counts().head(10))
```
